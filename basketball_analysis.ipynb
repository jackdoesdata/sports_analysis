```python
#create a list of each players dataframe
lst = [KobeBryant,JoeJohnson,LeBronJames,CarmeloAnthony,DwightHoward,ChrisBosh,ChrisPaul,KevinDurant,DerrickRose,DwayneWade]
```

## Ensure all data was in captured loop

```python
#created loop to insert each players stats into their dataframes
i = 0

while i < 10:
    df = lst[i]
    df['season'] = Seasons
    df['player'] = Players[i]
    df['salary'] = Salary[i]
    df['games'] = Games[i]
    df['mins_played'] = MinutesPlayed[i]
    df['field_goals'] = FieldGoals[i]
    df['field_goal_attempts'] = FieldGoalAttempts[i]
    df['points'] = Points[i]
    df['ppg'] = df['points']/df['games']
    df['fg_%'] = df['field_goals']/df['field_goal_attempts']
    df['mins_per_game'] = df['mins_played']/df['games']
    lst[i] = df
    i+=1

    
#union all dataframes
compiled_stats=pd.concat(lst, ignore_index=True)

#check that all players are included in unioned dataframe
compiled_stats.iloc[0:91:10]
```

# Perform analysis

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = 12,6
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet', 'darkslategrey', 'khaki', 'lightcoral']
```

### Kobe Bryant was the highest paid player from 2005 - 2014

```python
def plotter(data, Players):
    for i in Players:
        plt.plot(data[Pdict[i]], ls='--', marker='o')
    plt.xticks(list(range(0,10)), Seasons, rotation = 'vertical')
    plt.xlabel('Season')
    plt.legend(Players, loc='upper left', bbox_to_anchor=(1,1))
    plt.show()
plt.ylabel('Salary (Millions)')
plotter(Salary, Players)
```

```python
compiled_stats.groupby(['player'])['salary'].sum().plot('bar',color=colors)
plt.xticks(rotation='vertical')
plt.xlabel('Player')
plt.ylabel('2005-2014 Salary (millions)')
plt.title("Aggregate Salarys")
plt.show()
```

### As Kobe's Salary increased, his PPG was decreasing

```python
plt.ylabel('PPG')
plotter(Points/Games, ['KobeBryant'])
```

----

### Lebron James got a well deserved pay raise - every season he averaged over 25 PPG

```python
compiled_stats.groupby(['player'])['ppg'].mean().plot('bar', color=colors)
plt.title('Average Points Per Game')
plt.xlabel('Player')
plt.ylabel('Points')
plt.show()
```

```python
plt.ylabel('PPG')
plotter(Points/Games, ['LeBronJames'])
```

```python
plt.ylabel('Salary (millions)')
plotter(Salary, ['LeBronJames'])
```
